```python
import tqdm
with open("/home/khanhnd/youtube_crawler/KTSpeechCrawler/crawler/raw_data_6.txt", "w") as f:
    lines="wav_path, subtitle\n"
    index=0
    for wav in tqdm.tqdm(not_existed_wav):
        txt=not_existed_txt[index]
        try:
            with open(txt, "r") as f1:
                l=f1.readlines()[0].strip()
            lines=lines+wav+","+l+"\n"
        except:
            print("no_sub")
        index+=1
    f.write(lines)
```
