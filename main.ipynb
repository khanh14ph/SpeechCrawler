```python
with open("/home3/cuongld/asr-training/data/vocab/word_list_miss_indomain.txt", "r") as f:
    lst=f.readlines()
lst=[i.strip() for i in lst]
```

```python
lst
```

```python
import requests
import re
import os
import tqdm
import random
regex=r"(?<=watch\?v=)[\w]+(?=\")"
with open("link_list.txt", "a") as f:
    for j in tqdm.tqdm(lst):
        for i in range(1,2):  #search first n pages
            URL = "https://www.youtube.com/results?search_query="+j+"&sp=EgQIBCgB&page="+str(i)
            page = requests.get(URL)
            a=page.text
            match = re.findall(regex, a)
            l=random.choices(match, k=5)
            for t in l:
                link="https://www.youtube.com/watch?v="+t
                f.write(link+"\n")
            
```
